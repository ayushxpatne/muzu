```python
import pandas as pd
```

```python

classes = {
    'predict_month' : predict_month,
    'predict_months' : predict_months,
    'analyse_top_category' : analyse_top_category,
    'analyse_and_compare_categories' : analyse_and_compare_categories,
    'analyse' : analyse_expenses_questions,
    'advice' : advice,
    'greetings' : greetings

}

```

```python
def generate_class_wise_dataframe(key : str, value : list):
    df = pd.DataFrame(value, index=[i for i in range(len(value))],)
    df['class_'] = key
    return df
```

```python
dataframes = []
for key in classes:
    dataframes.append(generate_class_wise_dataframe( key=key, value=classes[key] ))
```

```python
concat_df = pd.concat(dataframes,).reset_index().drop(columns=['index']).rename(columns={
    0 : 'questions',
})
```

```python
dataset = concat_df
```

```python
categories = [
    'predict_month',
    'predict_months',
    'analyse_top_category',
    'analyse_and_compare_categories',
    'analyse',
    'advice'
    'greetings'
]
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(dataset, test_size=0.2)
```

```python
train
```

```python
# Importing necessary packages
from sklearn. feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
# Creating a model based on Multinomial Naive Bayes
model = make_pipeline(TfidfVectorizer(), MultinomialNB())
# Training the model with the train data
model.fit(train.questions, train.class_)
```

```python
import pickle

with open('naive_bayes_classification.pkl', 'wb') as file:
    pickle.dump(model, file)
```

```python
# Creating Labels for the test data
labels = model.predict(test.questions)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(test.class_, labels)
```

```python
# Creating confusion matrix and heat map
from sklearn.metrics import confusion_matrix

mat = confusion_matrix(test.class_, labels)
sns.heatmap(
    mat.T,
    square=True,
    annot=True,
    fmt="d",
    cbar=True,
    xticklabels=set(train.class_),
    yticklabels=set(train.class_),
)
# Plotting Heatmap of Confusion Matrix
plt.xlabel("true label")
plt.ylabel("predicted label")
```

```python
# Predicting category on new data based on trained model
def predict_class(s, model=model):
    pred= model.predict([s]) 
    return pred[0]
```
